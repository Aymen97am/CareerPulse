```python
from pydantic import BaseModel

class Person(BaseModel):
    name: str
    age: int
    
    model_config = {"extra": "allow"}

# Now this works
person = Person(name="John", age=30, useless_kwarg="extra", another_extra=123)
# print(person.extra_attributes)  # Doesn't store them in a special attribute
# But you can access them:
print(person.useless_kwarg)  # "extra"
print(person.another_extra)  # 123
```
